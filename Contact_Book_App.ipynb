```python
class Task:
    def __init__(self, name, contact, email, date):
        self.name = name
        self.contact = contact
        self.email = email
        self.date = date

    def __str__(self):
        return f"ID: {self.id} | Name : {self.name} | Contact: {self.contact} | Email: {self.email} | Date: {self.date}"

class TaskManager:
    def __init__(self):
        self.tasks = {}
        self.next_id = 1  # To generate unique task IDs

    def create_task(self, name, contact, email, date):
        task = Task(name, contact, email, date)
        task.id = self.next_id
        self.tasks[self.next_id] = task
        self.next_id += 1
        print(f"Contact {task.id} created successfully!")

    def read_all_tasks(self):
        if not self.tasks:
            print("No Contact available.")
        for task in self.tasks.values():
            print(task)

    def read_task(self, task_id):
        task = self.tasks.get(task_id)
        if task:
            print(task)
        else:
            print(f"Contact with ID {task_id} not found.")

    def update_task(self, task_id, name=None, contact=None, email=None, date=None):
        task = self.tasks.get(task_id)
        if task:
            if name:
                task.name = name
            if contact:
                task.contact = contact
            if email:
                task.email = email
            if date:
                task.date = date
            print(f"Contact {task_id} updated successfully!")
        else:
            print(f"Contact with ID {task_id} not found.")

    def delete_task(self, task_id):
        if task_id in self.tasks:
            del self.tasks[task_id]
            print(f"Contact {task_id} deleted successfully!")
        else:
            print(f"Contact with ID {task_id} not found.")

# Main program to demonstrate usage
if __name__ == "__main__":
    manager = TaskManager()
    manager.create_task("Vihan Lodhi", "1234567890","viham@gmail.com", "2025-03-30")
    manager.create_task("Rahul Singhaniya", "3456789234","rahul12@gmail.com", "2025-03-25")
    manager.create_task("Rina Mali", "1234567890","viham@gmail.com", "2025-03-30")
    manager.create_task("Vishakha Ravat", "3456789234","rahul12@gmail.com", "2025-03-25")
    manager.read_all_tasks()
    manager.update_task(2)
    manager.read_task(1)
    manager.delete_task(2)
    manager.read_all_tasks()



```
